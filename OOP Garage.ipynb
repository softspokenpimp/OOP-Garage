## Object Oriented Parking Garage:

Your assignment for today is to create a parking garage class to get more familiar with Object Oriented Programming(OOP).

Your parking garage class should have the following methods:
- takeTicket
- This should decrease the amount of tickets available by 1
- This should decrease the amount of parkingSpaces available by 1
- payForParking
- Display an input that waits for an amount from the user and store it in a variable
- If the payment variable is not empty then (meaning the ticket has been paid) -> display a message to the user that their  ticket has been paid and they have 15mins to leave
- This should update the "currentTicket" dictionary key "paid" to True
- leaveGarage
- If the ticket has been paid, display a message of "Thank You, have a nice day"
- If the ticket has not been paid, display an input prompt for payment
- Once paid, display message "Thank you, have a nice day!"
- Update parkingSpaces list to increase by 1 (meaning add to the parkingSpaces list)
- Update tickets list to increase by 1 (meaning add to the tickets list)

You will need a few attributes as well:
- tickets -> list
- parkingSpaces -> list
- currentTicket -> dictionary

```python
from IPython.display import clear_output

class Garage():
    
    def __init__ (self, tickets, parkingSpaces, currentTicket):
        #Using the idea that the range will be the amount of available spaces PLUS one
        # as this will enable the first index of "0" to signify that spaces are empty,
        #This enables appending no matter what, given the function in leaveGarage AND to show 
        #to user that the available spaces will be 0 at index 0. 
        self.tickets = list(range(11))
        self.parkingSpaces = list(range(11))
        self.currentTicket = {}
        
    def printTickets(self):
        print(f"Available tickets: {len(self.tickets)-1}")
        print(f"Available spaces: {len(self.parkingSpaces)-1}")
        
    def takeTicket(self):
        if len(self.tickets) == 1:
            print("Garage is full. Come back later.")
        else:
            self.currentTicket["Paid"] = False
            self.tickets.pop()
            self.parkingSpaces.pop()
        print(f"Available tickets: {len(self.tickets)-1}")
        print(f"Available spaces: {len(self.parkingSpaces)-1}")
        
    def payForParking(self):
        payment = input("Enter payment amount, pay whatever you want: ")
        try:
            if int(payment)<=0 :
                print("Please enter a valid amount.")
            else:
                print("Ticket has been paid, you have 15 minutes to vacate.") 
                self.currentTicket["Paid"] = True   
        except: 
            print("error")
        
    def leaveGarage(self):
        if self.currentTicket["Paid"] == True:
            print("Thank You, Have a nice day! :) ")
            self.currentTicket["Paid"] = False  #This disables one payment to allow for multiple leaves
        else:
            try:
                payment = input("Enter payment amount, pay whatever you want: ")
                if int(payment)<=0 :
                    print("Please enter a valid amount.")
                else:
                    print("Thank You, Have a nice day! :) ")
                    self.currentTicket["Paid"] = False
                self.tickets.append(self.tickets[-1]+1)
                self.parkingSpaces.append(self.parkingSpaces[-1]+1) 
                print(f"\nAvailable tickets: {len(self.tickets)-1}")
                print(f"Available spaces: {len(self.parkingSpaces)-1}")     
            except:
                print("error")


target_gar = Garage(list(range(10)),list(range(10)),{})


def run(): 
    while True: 
        ask = input("\n\nWhat do you want to do? \nT = Take a Ticket \nP = Pay for Parking \nL = Leave Grarage \nQ = Quit\n")
        if ask.lower() == "q":
            clear_output()
            target_gar.printTickets()
            print("Tansaction Cancelled")
            break
        elif ask.lower() == "t":
            clear_output()
            target_gar.takeTicket()
        elif ask.lower() == "p":
            clear_output()
            target_gar.payForParking()
        elif ask.lower() == "l":
            clear_output()
            target_gar.leaveGarage()
        else:
            clear_output()
            print("Try another command")


run()
```
